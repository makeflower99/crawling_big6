->채용공고단어의 빈도와 강의명 단어의 빈도는 독립아니다.:대립->연관이 있다
    
    검정결과 연관이 있다.(귀무가설기각)
    다만 어떤 연관이 있는지는 후속연구를 통해 판단해야 한다.
    모집단갯수 차이가 나는데 실제로 차이가 안난다는 것은 연관이 있다고 볼 수 없다.

정리:
    귀무가설을 기각했으므로 채용공고단어의 빈도와 강의명단어의 빈도는 독립이 아니다.
    다만 어떤 연관이 있는지는 후속연구를 통해 판단해야 한다.
    잔차시각화를 보면 연관이 크게 나는 단어가 있고 연관이 별로 없는 단어가 있다.
    이는 단어별로 연관의 정도가 다르고 이것이 카이제곱검정에서의 결과에 영향을 주었을 수 있다.
    하지만 전술하였듯 어떤연관이 있는지,연관의 정도파악은 후속연구를 통해 확인해야 한다.
    (모집단 갯수의 차이가 크게 나는데 분할표에서 빈도만 보고 연관이 있다 없다
     판단하는 것은 불가능,이거는 보고서쓸건 아니고 태클들어오면 반박용)
