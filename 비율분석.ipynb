```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

데이터리안 = pd.read_excel("데이터리안자료.xlsx",sheet_name = "asd")

데이터리안 = 데이터리안[['삭제후','빈도']]
데이터리안 = 데이터리안.loc[0:6,:]
데이터리안["비율"]=데이터리안["빈도"]/40

#이 데이터는 카이제곱에 쓴 그 데이터입니다.
data = {
    'SQL': [39*34.1, 93],
    'python,R': [24*34.1, 168+34],
    '가설설정 및 실험 설계(A/B 테스트)': [16*34.1, 62],#논문, 프로젝트 등의 가설설정과 실험설계가 필요한 파트
    '시각화': [12*34.1, 42],#시각화툴도 포함한다는 가정->근데 실제로 있지는 않았음
    "데이터 분석 방법론":[11*34.1,16],#딥러닝+머신러닝+자연어 등
    "통계적 역량":[15*34.1,4],#통계,수학,빅데이터등의 키워드
    "도메인 지식":[6*34.1,10],#금융,부동산 ,도매인등의 키워드
    
}
dff = pd.DataFrame(data, index=['채용공고', '강좌제목'])

강좌제목 = dff.iloc[1:2, :]
강좌제목비율 = 강좌제목/1365

채용공고 = dff.iloc[0:1, :]
채용공고비율 = (채용공고/34.1)/40


# 두 데이터프레임 생성
data1 = {
    'SQL': [0.068132],
    'python,R': [0.147985],
    '가설설정 및 실험 설계(A/B 테스트)': [0.045421],
    '시각화': [0.030769],
    '데이터 분석 방법론': [0.011722],
    '통계적 역량': [0.00293],
    '도메인 지식': [0.007326],
}

data2 = {
    'SQL': [0.975],
    'python,R': [0.6],
    '가설설정 및 실험 설계(A/B 테스트)': [0.4],
    '시각화': [0.3],
    '데이터 분석 방법론': [0.275],
    '통계적 역량': [0.375],
    '도메인 지식': [0.15],
}

df1 = pd.DataFrame(data1, index=['강좌제목'])
df2 = pd.DataFrame(data2, index=['채용공고'])

# 두 데이터프레임을 합치고 Transpose
merged_df = pd.concat([df1, df2]).T

# 꺾은선 그래프 그리기
plt.rc('font', family='Malgun Gothic') # 폰트 지정
plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정
plt.figure(figsize=(12, 6))
for col in merged_df.columns:
    plt.plot(merged_df.index, merged_df[col], marker='o', label=col)
#plt.grid()
plt.title('Comparison of 강좌제목 and 채용공고')
plt.xlabel('Category')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

# 채용공고와 강좌제목 키워드별 비교

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = {
    'SQL': [0.068132, 0.975],
    'python,R': [0.147985, 0.6],
    '가설설정 및 실험 설계(A/B 테스트)': [0.045421, 0.4],
    '시각화': [0.030769, 0.3],
    '데이터 분석 방법론': [0.011722, 0.275],
    '통계적 역량': [0.00293, 0.375],
    '도메인 지식': [0.007326, 0.15],
}

# 데이터프레임 생성
df = pd.DataFrame(data, index=['강좌제목', '채용공고'])

# 막대 그래프
fig, ax = plt.subplots(figsize=(12, 6))
width = 0.35  # 막대의 너비
categories = df.columns

# 강좌제목의 막대 그래프
bar1 = ax.bar(np.arange(len(categories)), df.loc['강좌제목'], width, label='강좌제목')

# 채용공고의 막대 그래프
bar2 = ax.bar(np.arange(len(categories)) + width, df.loc['채용공고'], width, label='채용공고')

# 레이블과 제목 추가
ax.set_ylabel('Ratio')
ax.set_title('Comparison of Ratios between Courses and Job Postings')
ax.set_xticks(np.arange(len(categories)) + width / 2)
ax.set_xticklabels(categories,rotation = 60)
ax.legend()

# 그래프 출력
plt.savefig('visualization.png', bbox_inches='tight') 
plt.show()

```
