```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time

driver = webdriver.Chrome()

driver.get('https://www.inflearn.com/courses/data-science')

ind = 1 
no = 1  
page = 1  
list1 = []  

while 1:
    try:
        # 이름, 가격
        name = driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{ind}]/div/a/div[2]/div[1]').text
        price = driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{ind}]/div/a/div[2]/div[4]').text    
        time.sleep(0.3)
        # 학생수
        driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{ind}]').click()
        try:
        # 학생수를 찾아서 저장
            students = driver.find_element(By.XPATH, f'//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[3]/span[2]/strong').text    
            list1.append([name, price, students])
        except NoSuchElementException:
        # 만약 학생수에 대한 XPath가 없을 경우, 0으로 처리
            list1.append([name, price, 0])
                
        driver.back() #뒤로가기
        # 클릭하고자하는 곳으로 스크롤해서 내려가기
        element = driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{ind}]')
        driver.execute_script("arguments[0].scrollIntoView(true);", element)
        time.sleep(1)
        ind += 1
        
    except:
        
        # 더보기 버튼 찾기
        if no == 1:
            driver.find_element(by='xpath',value = '//*[@id="courses_section"]/div/div/div/footer/nav/div/a').click()
            no += 1
            ind = 1
            page += 1
            continue
           
        if no < 20:
            driver.find_element(by='xpath',value = '//*[@id="courses_section"]/div/div/div/footer/nav/div/a[2]').click()
            no += 1
            ind = 1
            page += 1
            continue
        
        # 더이상 이동할수 없을 경우 종료
        else:
            break

pd.DataFrame(list1)
df_res = pd.DataFrame()

df_res = pd.concat([df_res, pd.DataFrame(list1)])
df_res.to_csv('./Inflearn_crawling.csv', index=False) # 엑셀 파일로 내보내기
```
